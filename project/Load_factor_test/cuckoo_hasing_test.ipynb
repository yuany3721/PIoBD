```python
import math
import random
import string
from Crypto.Cipher import AES
from binascii import b2a_hex, hexlify
```

#### Parameters

```python
ELEMENT_SIZE = pow(2,256)
SET_SIZE = pow(2,20)
ALPHABET = string.ascii_uppercase + string.digits
TRIALS = 100
CUCKOOHASH_TRIALS = 64
```

#### Utils

```python
def radnom256():
    return ''.join(random.choice(ALPHABET) for _ in range(32))
def utf8_to_int(s):
    return int(hexlify(s),16)
```

#### Generate the set

#### Hash Functions ( Hashed Values)

```python
mode = AES.MODE_ECB
key1 = radnom256()
key2 = radnom256()
key3 = radnom256()
    
cryptos1 = AES.new(key1.encode("utf-8"), mode)
cryptos2 = AES.new(key2.encode("utf-8"), mode)
cryptos3 = AES.new(key3.encode("utf-8"), mode)
```

#### Simple Hashing

```python
print("Simple Hashing")
sumedOccupancyRate = 0

for i in range (TRIALS):
    elementSet = set()
    while len(elementSet) < SET_SIZE:
        ra =  radnom256()
        elementSet.add(ra)
    elements = list(elementSet)
    
    hashes = []
    for element in elements:
        hashes += [utf8_to_int(cryptos1.encrypt(element.encode("utf-8")))]

    table = [-1 for _ in range(SET_SIZE)]

    cnt = 0
    for j in range(SET_SIZE):
        index = hashes[j] % SET_SIZE;
        if table[index] == -1:
            cnt += 1
            table[index] = elements[j]
    sumedOccupancyRate += cnt/SET_SIZE
    print("TRIAL #%d, occupancy rate: %f" %(i, cnt/SET_SIZE))
    
occupancyRate = sumedOccupancyRate/TRIALS
print("Average occupancy rate: ", occupancyRate)
```

#### Three-way Cuckoo Hashing

```python
def cuckoo_hashing_insert(table, hashes0, hashes1, hashes2, hashIndexes, elementIndex, hashIndex, trials):
    if trials >= CUCKOOHASH_TRIALS:
        return False
    hashes = hashes0 if hashIndex == 0 else (hashes1 if hashIndex == 1 else hashes2)
    index = hashes[elementIndex] % SET_SIZE;
    if table[index] == -1:
        table[index] = elementIndex
        hashIndexes[index] = hashIndex
        return True
    else:
        hi = hashIndexes[index]
        e = table[index]
        table[index] = elementIndex
        hashIndexes[index] = hashIndex
        success = cuckoo_hashing_insert(table, hashes0, hashes1, hashes2, hashIndexes, e, (hi+1)%3, trials+1)
        return success
    
        
sumedOccupancyRate = 0

for i in range (TRIALS):
    elementSet = set()
    while len(elementSet) < SET_SIZE:
        ra =  radnom256()
        elementSet.add(ra)
    elements = list(elementSet)
    
    hashes0 = []
    hashes1 = []
    hashes2 = []
    
    for element in elements:
        hashes0 += [utf8_to_int(cryptos1.encrypt(element.encode("utf-8")))]
        hashes1 += [utf8_to_int(cryptos2.encrypt(element.encode("utf-8")))]
        hashes2 += [utf8_to_int(cryptos3.encrypt(element.encode("utf-8")))]
    
    hashIndexes = [-1 for _ in range (SET_SIZE)]
    table = [-1 for _ in range(SET_SIZE)]
    cnt = 0
    for j in range (SET_SIZE):
        success = cuckoo_hashing_insert(table, hashes0, hashes1, hashes2, hashIndexes, j, 0, 0)
        if success:
            cnt += 1
    sumedOccupancyRate += cnt/SET_SIZE
    print("TRIAL #%d, occupancy rate: %f" %(i, cnt/SET_SIZE))
    
occupancyRate = sumedOccupancyRate/TRIALS
print("Average occupancy rate: ", occupancyRate)
```

#### Two-way Cuckoo Hashing with Two Buckets

```python
def cuckoo_hashing_insert(table0, table1, lastEvicted, hashes0, hashes1, hashIndexes, elementIndex, hashIndex, trials):
    if trials > CUCKOOHASH_TRIALS:
        return False
    
    hashes = hashes0 if hashIndex == 0 else hashes1
    index = hashes[elementIndex] % (SET_SIZE//2);
    table = table0 if lastEvicted[index] == 0 else table1
    lastEvicted[index] = (lastEvicted[index]+1) % 2
        
    if table[index] == -1:
        table[index] = elementIndex
        hashIndexes[index] = hashIndex
        return True
    else:
        hi = hashIndexes[index]
        e = table[index]
        table[index] = elementIndex
        hashIndexes[index] = hashIndex
        success = cuckoo_hashing_insert(table0, table1, lastEvicted, hashes0, hashes1, hashIndexes, e, (hi+1)%2, trials+1)
        return success

sumedOccupancyRate = 0
for i in range (TRIALS):
    elementSet = set()
    while len(elementSet) < SET_SIZE:
        ra =  radnom256()
        elementSet.add(ra)
    elements = list(elementSet)
    
    hashes0 = []
    hashes1 = []
    
    for element in elements:
        hashes0 += [utf8_to_int(cryptos1.encrypt(element.encode("utf-8")))]
        hashes1 += [utf8_to_int(cryptos2.encrypt(element.encode("utf-8")))]
    
    hashIndexes = [-1 for _ in range (SET_SIZE)]
    table0 = [-1 for _ in range(SET_SIZE//2)]
    table1 = [-1 for _ in range(SET_SIZE//2)]
    lastEvicted = [0 for _ in range(SET_SIZE//2)]
    cnt = 0
    for j in range (SET_SIZE):
        success = cuckoo_hashing_insert(table0, table1,lastEvicted, hashes0, hashes1, hashIndexes, j, 0, 0)
        if success:
            cnt += 1
    sumedOccupancyRate += cnt/SET_SIZE
    print("TRIAL #%d, occupancy rate: %f" %(i, cnt/SET_SIZE))
    
occupancyRate = sumedOccupancyRate/TRIALS
print("Average occupancy rate: ", occupancyRate)
```

#### Three-way Cuckoo Hashing Using Separete Tables

```python
def cuckoo_hashing_insert(table0, table1, table2, hashes0, hashes1, hashes2, elementIndex, hashIndex, trials):
    if trials >= CUCKOOHASH_TRIALS:
        return False
    hashes = hashes0 if hashIndex == 0 else (hashes1 if hashIndex == 1 else hashes2)
    table = table0 if hashIndex == 0 else (table1 if hashIndex == 1 else table2)
    index = hashes[elementIndex] % (SET_SIZE//3);
    if table[index] == -1:
        table[index] = elementIndex
        hashIndexes[index] = hashIndex
        return True
    else:
        hi = hashIndexes[index]
        e = table[index]
        table[index] = elementIndex
        hashIndexes[index] = hashIndex
        success = cuckoo_hashing_insert(table0, table1, table2, hashes0, hashes1, hashes2, e, (hashIndex+1)%3, trials+1)
        return success
    
        
sumedOccupancyRate = 0

for i in range (TRIALS):
    elementSet = set()
    while len(elementSet) < SET_SIZE:
        ra =  radnom256()
        elementSet.add(ra)
    elements = list(elementSet)
    
    hashes0 = []
    hashes1 = []
    hashes2 = []
    
    for element in elements:
        hashes0 += [utf8_to_int(cryptos1.encrypt(element.encode("utf-8")))]
        hashes1 += [utf8_to_int(cryptos2.encrypt(element.encode("utf-8")))]
        hashes2 += [utf8_to_int(cryptos3.encrypt(element.encode("utf-8")))]
    
    table0 = [-1 for _ in range(SET_SIZE//3)]
    table1 = [-1 for _ in range(SET_SIZE//3)]
    table2 = [-1 for _ in range(SET_SIZE//3)]
    cnt = 0
    for j in range (SET_SIZE):
        success = cuckoo_hashing_insert(table0, table1, table2, hashes0, hashes1, hashes2, j, 0, 0)
        if success:
            cnt += 1
    sumedOccupancyRate += cnt/SET_SIZE
    print("TRIAL #%d, occupancy rate: %f" %(i, cnt/SET_SIZE))
    
occupancyRate = sumedOccupancyRate/TRIALS
print("Average occupancy rate: ", occupancyRate)
```
